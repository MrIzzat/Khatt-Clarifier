```python
def rotate_bound(image, angle):
    # grab the dimensions of the image and then determine the
    # center
    (h, w) = image.shape[:2]
    (cX, cY) = (w // 2, h // 2)
    # grab the rotation matrix (applying the negative of the
    # angle to rotate clockwise), then grab the sine and cosine
    # (i.e., the rotation components of the matrix)
    M = cv.getRotationMatrix2D((cX, cY), -angle, 1.0)
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1])
    # compute the new bounding dimensions of the image
    nW = int((h * sin) + (w * cos))
    nH = int((h * cos) + (w * sin))
    # adjust the rotation matrix to take into account translation
    M[0, 2] += (nW / 2) - cX
    M[1, 2] += (nH / 2) - cY
    # perform the actual rotation and return the image
    return cv.warpAffine(image, M, (nW, nH))
```

# Merging all the augmented and non-augmented data into a single directory

## Creating the necessary directories

```python
if not os.path.exists("./Full Dataset"):
  os.mkdir("./Full Dataset")

if not os.path.exists("./Full Dataset/Labels"):
  os.mkdir("./Full Dataset/Labels")

if not os.path.exists("./Full Dataset/Images"):
  os.mkdir("./Full Dataset/Images")
```

## Getting directories of base and destination paths

```python
base_imgaug = "./img_aug/"
base_labelaug= "./labels_aug/"

base_imgaug2 = "./img_aug2/"
base_labelaug2= "./labels_aug2/"

base_img = "./Dataset/Total Images/"
base_label = "./Dataset/Total GT/"

destination_img = "./Full Dataset/Images/"
destination_label= "./Full Dataset/Labels/"

basePaths_images = [base_imgaug,base_img,base_imgaug2]
basePaths_labels = [base_labelaug,base_label,base_labelaug2]
```

## Do the moving Process

```python

for basePath in basePaths_images:
  for path in sorted(os.listdir(basePath)):
    shutil.move(basePath+path, destination_img+path)

for basePath in basePaths_labels:
  for path in sorted(os.listdir(basePath)):
    shutil.move(basePath+path, destination_label+path)
```
